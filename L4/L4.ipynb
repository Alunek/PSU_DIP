```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt


image1 = cv.imread('img.jpg')
rgb_image1 = cv.cvtColor(image1, cv.COLOR_BGR2RGB)
hsv_image1 = cv.cvtColor(image1, cv.COLOR_BGR2HSV)
gray_image1 = cv.cvtColor(image1, cv.COLOR_BGR2GRAY)

channels = [0]
histSize = [256]
range = [0, 256]

gs = plt.GridSpec(2, 2)
plt.figure(figsize=(10, 8))
plt.subplot(gs[0])
plt.imshow(gray_image1, cmap='gray')
plt.xticks([]), plt.yticks([])
plt.subplot(gs[1])
plt.hist(gray_image1.reshape(-1), 256, range)
plt.show()
```

```python
image = gray_image1

ret, thresh1 = cv.threshold(image, 125, 255, cv.THRESH_BINARY)
ret, thresh2 = cv.threshold(image, 123, 255, cv.THRESH_BINARY)
ret, thresh3 = cv.threshold(image, 120, 255, cv.THRESH_BINARY)

gs = plt.GridSpec(2, 2)
plt.figure(figsize=(100, 20))
plt.subplot(gs[0])
plt.imshow(gray_image1, cmap='gray')
plt.subplot(gs[1])
plt.title("порог 125")
plt.imshow(thresh1, cmap='gray')
plt.subplot(gs[2])
plt.title("порог 123")
plt.imshow(thresh2, cmap='gray')
plt.subplot(gs[3])
plt.title("порог 120")
plt.imshow(thresh3, cmap='gray')
plt.show()

#Перевод: Начните делать все, что можете, даже то, о чем можете хотя бы мечтать.
```
